```python
import matplotlib.pyplot as plt 
from matplotlib import style
import pandas as pd 
```

```python
# data sets link: https://www.kaggle.com/datasets/lava18/google-play-store-apps
df_google_play_store_apps = pd.read_csv('C:\\Users\\Acer\\Downloads\\archive\\googleplaystore.csv')
df_google_play_store_apps.shape
```

```python
df_google_play_store_apps = pd.read_csv('C:\\Users\\Acer\\Downloads\\archive\\googleplaystore.csv',nrows=1000)
df_google_play_store_apps.shape
```

```python
df_google_play_store_apps.head()
```

```python
x = df_google_play_store_apps["Rating"]
y = df_google_play_store_apps["Reviews"]
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y)
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
# from above scatter plot : we came to know that :->
# most of the people give rating range between  4.0 to 4.7 in the reviews on the google apps 
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='r',marker='*',s=100,alpha=0.5,linewidth=8,
           edgecolor='g')
#marker='o','+','.','_','<','>','^',',','*'
# s = 100 means--> marker size 
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='r',marker='*',s=100,alpha=0.5,linewidth=10,
           edgecolor='g')
plt.scatter(x,df_google_play_store_apps['Installs'],c='y',marker='o',s=100,alpha=0.5,
           linewidth=10,edgecolor='c')

#marker='o','+','.','_','<','>','^',',','*'
# s = 100 means--> marker size 
plt.title("Google Play Store Apps Scatter Plot",fontsize=18)
plt.xlabel("Rating",fontsize=15)
plt.ylabel("Reviews and Installs",fontsize=15)
plt.show()
```
