```python
from matplotlib import pyplot as plt 
```

```python
x = [1,2,3,4]
y = [2,4,6,8]
plt.plot(x,y)
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
c = ['r', 'y', 'g', 'b']

plt.bar(x, y, color=c)
plt.show()
```

```python
# bar plot 
```

```python
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.xlabel('language',fontsize=15)
plt.ylabel('popularity',fontsize=15)
plt.title('LEVEL',fontsize=15)
c = ['y','b','m','g']
plt.bar(x,y,width=0.3,color=c,align='center')

plt.show()
```
