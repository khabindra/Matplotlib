How to save figure (plots and charts) ? 

```python
import matplotlib.pyplot as plt
```

```python
plt.pie([40,30,20])
plt.savefig('Pie_char')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.pie([40, 30, 20])
plt.savefig('Pie_chart.png', dpi=1000, facecolor='green')
#dpi 
plt.show()
```

```python
plt.pie([40, 30, 20])
plt.savefig('Pie_chart.png', dpi=100, facecolor='green')
# In png format quality parameter is not applicable .
plt.show()
```
