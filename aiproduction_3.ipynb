```python
import matplotlib.pyplot as plt 
import numpy as np 
from matplotlib import style
```

```python
# Dataset of 'Artificial Intelligence (AI) Class" 
#classes = ["Python", "R", "Artificial Intelligence", "Machine Learning", "Data Science"]
 
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 10, 20, 25, 10] # out of 100 student in each class
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]
```

```python
plt.bar(classes,class1_students)
```

```python
plt.barh(classes,class1_students)
```

```python
plt.bar(classes,class1_students,width=0.2,align="edge",color="y",
       edgecolor="m",linewidth=5,alpha=0.8,linestyle="--",
       label=" Class 1 Students")
plt.legend(loc=1)
plt.show()
```

```python
plt.bar(classes,class1_students,width=0.2,align="edge",color="y",
       edgecolor="m",linewidth=5,alpha=0.8,linestyle="--",
       label=" Class 1 Students",visible=False)
plt.legend(loc=1)
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,width=0.6,align="edge",color="k",
       edgecolor="m",linewidth=5,alpha=0.8,linestyle="--",
       label=" Class 1 Students")
plt.legend(loc=1)
plt.show()
```

```python
# for plotting the same bar graph horizontally.
plt.figure(figsize=(16,9))
plt.barh(classes,class1_students,align="edge",color="k",
       edgecolor="m",linewidth=5,alpha=0.8,linestyle="--",
       label=" Class 1 Students")
plt.legend(loc=1)
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,align="edge",color="k",
       edgecolor="m",linewidth=5,alpha=0.8,linestyle="--",
       label=" Class 1 Students")
plt.title("Bar Chart of AI Class",fontsize=18)
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of Students",fontsize=15)
plt.legend(loc=1)
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,width=0.2,color="b",
        label=" Class 1 Students")

plt.bar(classes,class2_students,width=0.2,color="g",
        label=" Class 2 Students")

plt.title("Bar Chart of AI Class",fontsize=18)
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of Students",fontsize=15)
plt.legend(loc=1)
plt.show()
# this method overlaps the data bar chart of each other . so we reject this method .
```

```python
plt.figure(figsize=(16,9))

class_index = np.arange(len(classes))
print(class_index)
width = 0.2
plt.bar(class_index,class1_students,width,color="b",
        label=" Class 1 Students")

plt.bar(class_index+width,class2_students,width,color="g",
        label=" Class 2 Students")

plt.xticks(class_index+width,classes,rotation=25,fontsize=20)

plt.title("Bar Chart of AI Class",fontsize=18)
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of Students",fontsize=15)
plt.legend(loc=1)
plt.show()
```

```python
plt.figure(figsize=(16,9))

class_index = np.arange(len(classes))
print(class_index)
width = 0.2
plt.bar(class_index,class1_students,width,color="b",
        label=" Class 1 Students")

plt.bar(class_index+width,class2_students,width,color="g",
        label=" Class 2 Students")

plt.bar(class_index+width+width,class3_students,width,color="y",
        label=" Class 3 Students")


plt.xticks(class_index+width,classes,rotation=25,fontsize=20)

plt.title("Bar Chart of AI Class",fontsize=18)
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of Students",fontsize=15)
plt.legend(loc=1)
plt.show()
```

```python
# for horizontal bar chart 
plt.figure(figsize=(16,9))

class_index = np.arange(len(classes))
print(class_index)
width = 0.2
plt.barh(class_index,class1_students,width,color="b",
        label=" Class 1 Students")

plt.barh(class_index+width,class2_students,width,color="g",
        label=" Class 2 Students")

plt.barh(class_index+width+width,class3_students,width,color="y",
        label=" Class 3 Students")


plt.yticks(class_index+width,classes,rotation=25,fontsize=20)

plt.title("Bar Chart of AI Class",fontsize=18)
plt.ylabel("Classes",fontsize=15)
plt.xlabel("No. of Students",fontsize=15)
plt.legend(loc=1)
plt.show()
```
